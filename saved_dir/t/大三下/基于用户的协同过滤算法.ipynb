### 基础算法

#### 找到和目标用户兴趣相似的用户集合
** 利用行为的相似度计算兴趣的相似度

*实现余弦相似度

```python
def user_similarity(train):
    import math
    
    w = dict()
    for u in train.keys():
        for v in train.keys():
            if u == v:
                continue
            uv = u + '|' + v
            w[uv] = len(set(u2_items[u]) & set(u2_items[v]))
            w[uv] = w[uv] / math.sqrt(len(train[u]) * len(train[v]))
            
    return w

# print(user_similarity(u2_items))
```
